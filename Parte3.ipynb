```python
import matplotlib.pyplot as plt
import pydicom

ds = pydicom.dcmread('D0120.dcm',force=True)
img = ds.pixel_array
img = img/img.max()

plt.imshow(img, cmap="gray")
```

```python
from skimage import feature, segmentation
import numpy as np
import matplotlib
matplotlib.use('TkAgg') 
```

```python

plt.ion()
plt.title("Seleccione puntos con tecla izquierda, \n borre el último punto con tecla derecha,\n finalize con tecla central", fontsize=16)
plt.imshow(img, cmap = 'gray')
plt.pause(0.001)
seed_coord = plt.ginput(-1)
plt.close()

print(seed_coord)
```

```python
lista = []
    
for coord in seed_coord:
    lista.append(coord)    
array = np.array(lista)  
snake = array.reshape(len(seed_coord),2)

matrix = np.zeros(snake.shape)
matrix[:,1] = snake[:,0]
matrix[:,0] = snake[:,1]
print(matrix)
```

```python
contornos = segmentation.active_contour(img,snake)
plt.imshow(contornos,cmap='gray')
```
