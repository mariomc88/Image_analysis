#### CONCLUSIÓN SOBRE LOS RESULTADOS

#### Algoritmo de contorno activo (snake)

El algoritmo de contorno activo o **Snake**  utiliza fuerzas de energía y restricciones para definir el área de segmentación. La aplicación principal de contornos activos en el procesamiento de imágenes es definir una forma suave en la imagen y formar un contorno cerrado para la región. Existen distintos tipos como el modelo de serpiente, el modelo de serpiente de flujo vectorial, el modelo de globo y contornos geodésicos o geométricos. Se utilizan bastante en imagen médica.

#### Evaluación del algoritmo snakes en las imágenes

Tras los resultados obtenidos podemos concluir que snakes sin ninguna mejora adicional da muy buenos resultados en seguimiento de bordes para la segmentación de regiones de interés. Sin embargo, snakes es un algoritmo algo inestable ya que depende de que se introduzcan los parámetros adecuados para que se haga un seguimiento de contornos perfecto. Si no se encuentran los parámetros adecuados puede deformar mucho el borde del de referencia como se puede observar en la imagen de la columna vertebral. 
Algunos de estos parámetros pueden ser:
- El uso de círculo o elipse
- El radio del círculo o la elipse
- Posición del centro del círculo o elipse.
