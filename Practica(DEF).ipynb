###### Importamos los módulos necesarios y las funciones creadas para la adición de ruido y filtrado de la imagen. Importamos la imagen, la normalizamos entre 0 y 1 y la representamos

```python
from matplotlib.pylab import plt
import numpy as np
import pydicom
from IPython.display import clear_output
from funciones import gaussian_noise, impulsive_noise, both_noises, non_local, non_local_mejorado, filtro_anisotropico, imgplot
```

```python
ds = pydicom.dcmread("D0120.dcm", force=True) 
img = np.array(ds.pixel_array)/ds.pixel_array.max() #Normalizamos la imagen original de 0 a 1.
imgplot(img);
```

###### Añadimos ruido a la imagen original puesto que es una imagen ya procesada. Se añadirán los ruidos gaussiano e impulsivo.

* En la siguiente celda añadimos el ruido gaussiano a la imagen y representamos la imagen original frente a la ruidosa

```python
gaussian_img = gaussian_noise(img) # gaussian_noise(image, std = 0.05)

imgplot(img, gaussian_img)
```

* En la siguiente celda añadimos el ruido impulsivo a la imagen y representamos la imagen original frente a la ruidosa

```python
impulsive_img = impulsive_noise(img) # impulsive_noise(image, amount = 0.05)

imgplot(img, impulsive_img)
```

* En la siguiente celda añadimos los ruidos gaussianos e impulsivo a la imagen y representamos la imagen original frente a la ruidosa. Normalizamos también la imagen ruidosa.

```python
noisy_img = both_noises(img) # both_noises(image, std = 0.05, amount = 0.05)

imgplot(img, noisy_img)
```

###### Aplicamos el filtro Non-Local-Means (NLM) sin mejoras y sin paralelización por lo que su ejecución será extremadamente lenta.

* Representamos la imagen ruidosa junto a la filtrada.

```python
NL_filtered = non_local(noisy_img, 0.15) # non_local(image,h)

imgplot(noisy_img, NL_filtered)
```

###### Aplicamos el filtro Non-Local-Means (NLM) con mejoras y con paralelización por lo que la ejecución del código será rápida y será posible aplicar el filtrado a la imagen completa de 512x512.

* NL_filtered1 es un ejemplo de imagen sin mejoras de NLM implementadas y NL_filtered2 es un ejemplo de imagen con mejoras de NLM implementadas.
* Representamos la imagen ruidosa junto a las imágenes filtradas.

```python
NL_filtered1 = non_local_mejorado(noisy_img, 0.15)    # non_local_mejorado(image,h,CPP = False,pond_orig = False,D=0,alpha=0)
clear_output()
```

```python
NL_filtered2 = non_local_mejorado(noisy_img, 0.15, CPP = True,pond_orig = True, D = 0.10, alpha = 4) 
clear_output()
```

```python
imgplot(noisy_img, NL_filtered1, NL_filtered2)
```

###### Comparamos con la funcion  *denoise_nl_means* del paquete skimage.restoration.

```python
from skimage.restoration import denoise_nl_means, estimate_sigma

sigma_est = estimate_sigma(noisy_img, multichannel=False)
denoised = denoise_nl_means(noisy_img, patch_size=3, patch_distance=100, h=0.15, sigma = sigma_est)
imgplot(img, denoised, NL_filtered1, NL_filtered2)
```

```python
imgplot(img, denoised)
```

###### Aplicamos un filtro anisotrópico a la imagen completa.

```python
A_filtered = filtro_anisotropico(img, iter = 10) # filtro_anisotropico(img, treshold = 0.3, size_filter = 3, iter = 3)

imgplot(img, A_filtered)
```

###### Comparamos con la función de Anisotropic_Diffuision_ref.py

```python
from Anisotropic_Diffusion_ref import anisodiff
filtered_image = anisodiff(img, niter = 30)

imgplot(noisy_img, filtered_image, A_filtered)
```
